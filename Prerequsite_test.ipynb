1) (k) x (n x m) multiplications and (k-1) x (n x m) additions we need to perform a matrix multiplication between a (n, k) and (k, m) matrix

the process for matrix multiplication is to multiply elements of a row from first matrix to the element of second matrix, as in any row of first matrix there will be 'k' elements and any column of second matrix has 'k' elements, so we multiply 'k' times to find any element of resultant matrix. 
To find all elements(i.e., n x m) of resultant matrix, we will perform (k) x (n x m) multiplications.
and in while finding single element we will add all those multiplied terms, so in this case we will add 'k-1' times.
To find all elements(i.e., n x m) of resultant matrix, we will perform (k-1) x (n x m) additions.


```python
# 2)
import time
import numpy as np
import random

def matrix_multiplication(a, b):
    r1, c1, r2, c2 = len(a), len(a[0]), len(b), len(b[0])
    
    if c1 != r2:
        return
    
    multiplied_matrix = []
    
    for i in range(r1):
        row_of_multiplied_matrix = []

        for j in range(c2):
            item = 0

            for k in range(c1):
                item = item + a[i][k]*b[k][j]

            row_of_multiplied_matrix.append(item)
        
        multiplied_matrix.append(row_of_multiplied_matrix)
    
    return multiplied_matrix


# A = [
#     [2, 1, 3, 4],
#     [0, 5, 2, 1],
#     [8, 2, 4, 6]
# ]
# B = [
#     [7, 2, 1, 5, 3],
#     [4, 0, 9, 8, 2],
#     [6, 1, 3, 7, 4],
#     [2, 5, 8, 1, 6]
# ]

A = np.random.rand(100, 100)
B = np.random.rand(100, 100)

start_time = time.time()
output = matrix_multiplication(A, B)
my_duration = time.time() - start_time

print(output)
```

```python
import numpy as np

start_time = time.time()
output_ = np.matmul(A, B)
numpy_duration = time.time() - start_time

print(output_)
```

```python
print("Time taken by the function I defined:", my_duration)
print("Time taken by the function in numpy library:", numpy_duration)
```

4) Gradient of (x^2).y + (y^3).(sin(x)) is [ 2.x.y + (y^3)(cos(x))  , (x^2) + (3.(y^2).(sin(x)) ]

```python
# 5)

import jax
import jax.numpy as jnp
import math

# Define a mathematical equation as a function
def equation(x, y):
    return ((x**2) * y) + ((y**3) * (jax.numpy.sin(x)))

# Create JAX arrays as input
x, y = jnp.array([math.pi, 1.0])

# Compute the gradient of the equation with respect to the input array
grad_equation = jax.grad(equation, argnums=(0, 1))(x, y)

print("Gradient of the equation:", grad_equation)
```

```python
# 13)

def arg_min(lst):
    min = lst[0]
    min_ind = 0
    for i in range(1, len(lst)):
        if lst[i]<min:
            min = lst[i]
            min_ind = i
            
    return min_ind

lst = np.array([5, 6, 8, 1, 9, 2])
arg_min(lst)
```
